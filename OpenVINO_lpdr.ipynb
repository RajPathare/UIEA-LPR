```
#import necessary libraries
import cv2
import numpy as np
#import colab's imshow
from google.colab.patches import cv2_imshow
```

```
#read the image
vpd = cv2.imread('car_1.bmp')
vpd.shape
```

```
#display the image
cv2_imshow(vpd)
```

```
# replicate the output received from vpd.py
out = [0.        , 2.,         0.9430796 , 0.44141868, 0.64430356 ,0.5247948,
 0.68497896]

```

```
# create a copy of the image
frame = vpd.copy()
frame.shape
```

```
# resize the image
frame = cv2.resize(frame, (300,300))
frame.shape
```

```
# calculate the extreme co-ordinates of the bounding boxes
xmin = int(out[3] * frame.shape[0])
ymin = int(out[4] * frame.shape[1])
xmax = int(out[5] * frame.shape[0])
ymax = int(out[6] * frame.shape[1])

print(xmin ,ymin, xmax, ymax)
```

```
# draw the bounding box around the output
cv2.rectangle(frame, (xmin, ymin), (xmax, ymax), color=(0, 255, 0))
```

```
# display the output
cv2_imshow(frame)
```

```
# write the image
cv2.imwrite("CAR-PLATE-output.jpg", frame)
```

```
clone = vpd.copy()
clone = cv2.resize(clone, (300, 300))
clone.shape
```

```
# slice the image to extract the output
extracted = frame[ymin:ymax, xmin:xmax]
```

```
cv2_imshow(extracted)
```

```
# finally write the desired output
cv2.imwrite('extracted_lp.jpg', extracted)
```
